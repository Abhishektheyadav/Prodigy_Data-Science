```python
plt.figure(figsize=(12,6))
# Create a bar chart
plt.bar(data['Country Name'], data['2021'], color='purple')

# Add labels and title
plt.xlabel('Countries(1960)')
plt.ylabel('Populations(1960)')
plt.title('Bar Chart for Your Dataset')
# plt.figure(figsize=(20,20))
plt.xticks(rotation=90)
# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'

# Extract data for India
India_data = df[df['Country Name'] == 'India']

# Extract data for China
China_data = df[df['Country Name'] == 'China']

# Set the figure size (width, height) in inches
plt.figure(figsize=(12, 8))

# Plot the population trends for India
plt.plot(India_data.columns[4:], India_data.iloc[0,4:], label='India', marker='o')

# Plot the population trends for China
plt.plot(China_data.columns[4:],China_data.iloc[0,4:], label='China', marker='o')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Trends Comparison Between India and China')
plt.xticks(rotation=90)

# Add legend
plt.legend()

# Display the chart
plt.show()
```
